```python
import pandas as pd
import glob
```

```python
# List all CSV files in the directory
csv_files = glob.glob("dataFiles/*.csv")
```

```python
# Read each CSV into a DataFrame
df_list = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames (headers are not duplicated)
merged_df = pd.concat(df_list, ignore_index=True)
```

```python
merged_df.sort_values(by=['S.no'], inplace=True)
```

```python
# Save to a new CSV (header will appear only once)
merged_df.to_csv("dataFiles/merged_dataframes.csv", index=False)
```
